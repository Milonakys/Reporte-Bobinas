```python
from math import*
import numpy as np
def calculo():

        ############
    #########################################################################
    ########################################################################
    #from math import*
    #import numpy as np
    #import drawSvg as draw

    #########################################################################
    ########################################################################
    paso= float(e1.get())
    R=float(e2.get()) #Radio del estator en [mm]
    b=float(e9.get()) # ancho de bobina
    h=float(e3.get()) # altura de bobina
    nu=float(e4.get())*pi/180 #angulo nudillo
    ranuras = int(e5.get()) #ranuras estator
    dpbob = float(e6.get()) ##Distancia piernas bobina en el nudo
    pbr= float(e7.get())  # Profundidad bajo ranura
    pbc =float(e8.get()) # Profundidad bajo cuña
    a=pbc # alto de 2 bobinas en una ranura
    c= pbr-pbc #espacio para la cuña
    anchc = float(e10.get()) # ancho cuña
    theta= (((paso-1)*360/ranuras)*pi)/180 # paso polar en [°]
    
    #########################################################################
    ########################################################################
    #theta= (80*pi)/180 # paso polar en [°]
    #R=236 #Radio del estator en [mm]
    #b=11 # ancho de bobina
    #h=125# altura de bobina
    #nu=19*pi/180 #angulo nudillo
    #ranuras = 72 #ranuras estator
    #dpbob = 12.5 ##Distancia piernas bobina en el nudo
    #pbr= 60  # Profundidad bajo ranura
    #pbc =56 # Profundidad bajo cuña
    #a=pbc # alto de 2 bobinas en una ranura
    #c=pbr-pbc #espacio para la cuña
    #anchc = 14 # ancho cuña
    #########################################################################
    ########################################################################


    angpm = atan((0.5*b)/(R))*180/pi #Angulo entre la punta de la ranura y el medio de esta
    angxranura=360/ranuras
    phi=(pi/2)-(theta/2)
    alpha=0.5*pi-nu
    chi=atan((b/2)/(R+c+(a/2)))
    H=((R+c+(a/2))/cos(chi))
    y=2*sin(((theta)/2)+chi)*H
    z=sqrt((y**2)+((a/2)**2)-2*(a/2)*y*cos(phi))
    gamma=asin(((a/2)*sin(phi))/(z))
    Phi1= phi-gamma
    Phi2=phi+gamma 
    print('El ángulo Phi1== %.2f ° , y el angulo Phi2== %.2f ° ' % ((Phi1*180/pi),(Phi2*180/pi)));
    #########################################################################
    ########################################################################

    chi=atan((b/2)/(R+c+(a/2)))
    HX=((R+c+(a/2))/cos(chi))
    yX=2*sin(((theta)/2)-chi)*H
    X1=sqrt((yX**2)+((a/2)**2)-2*(a/2)*yX*cos(phi))

    chi2=atan((b/2)/(R+c+(a)))
    HX2=((R+c+(a))/cos(chi2))
    yX2=2*sin(((theta)/2)+chi2)*HX2
    X2=sqrt((yX2**2)+((a)**2)-2*(a)*yX2*cos(phi))
    print('La distancia X1 == %.2f [mm]  ' % (X1))
    print('La distancia X2 == %.2f [mm]  ' % (X2))
    #########################################################################
    ########################################################################


    chix=atan((b/2)/(R+c))
    chiy=atan((b/2)/(R+c+a))
    d=sqrt(((a+dpbob)**2)+b**2)
    beta=acos((a+dpbob)/d)
    K= ((R+c)*cos((theta/2)+chix)+(R+c+a/2)*cos((theta/2)+chix))/(2)+h*cos(gamma)
    L=d*sin(alpha+beta)
    Xc=K-L
    s1=0.5*d*tan(nu-beta)
    s2=L*1.2*tan(gamma)
    Yc=s1-s2
    Xr=(R+c)*cos(0.5*theta+chix)
    Yr=(R+c)*sin(0.5*theta+chix)
    Xt=R*cos((0.5*theta+chix)/2)
    Yt=R*sin((0.5*theta+chix)/2)
    ##############################
    x= [Xc,Xr]
    y= [Yc,Yr]
    coefficients = np.polyfit(x,y,1)
    A = coefficients [0]
    B=-1
    C= coefficients [1]
    distancia= abs(A*Xt+B*Yt+C)/sqrt((A**2)+(B**2))
    print('La distancia máxima entre la bobina y el núcleo == %.3f [mm]  ' % (distancia))
    #########################################################################
    ########################################################################

    PuntoCfinal= [x[0]*cos(-pi/2) - y[0]*sin(-pi/2), x[0]*sin(-pi/2) + y[0]*cos(-pi/2)]
    PuntoRfinal= [x[1]*cos(-pi/2) - y[1]*sin(-pi/2), x[1]*sin(-pi/2) + y[1]*cos(-pi/2)]
    Pa1=[(R+c+0.5*a)*cos(0.5*theta+chi),-(R+c+0.5*a)*sin(0.5*theta+chi)]
    Pa2=[(R+c+0.5*a)*cos(0.5*theta-chi),-(R+c+0.5*a)*sin(0.5*theta-chi)]
    Pa3=[(R+a+c)*cos(0.5*theta-chiy),-(R+a+c)*sin(0.5*theta-chiy)]
    Pa4=[(R+a+c)*cos(0.5*theta+chiy),-(R+a+c)*sin(0.5*theta+chiy)]
    #########
    Pa1f=[Pa1[1],-Pa1[0]]
    Pa2f=[Pa2[1],-Pa2[0]]
    Pa3f=[Pa3[1],-Pa3[0]]
    Pa4f=[Pa4[1],-Pa4[0]]
    #########
    Pr1=[Xr,Yr]
    Pr2=[(R+c)*cos(0.5*theta-chix),(R+c)*sin(0.5*theta-chix)]
    Pr3=[(R+0.5*a+c)*cos(0.5*theta-chi),(R+0.5*a+c)*sin(0.5*theta-chi)]
    Pr4=[(R+0.5*a+c)*cos(0.5*theta+chi),(R+0.5*a+c)*sin(0.5*theta+chi)]
    ##########
    Pr1f=[Pr1[1],-Pr1[0]]
    Pr2f=[Pr2[1],-Pr2[0]]
    Pr3f=[Pr3[1],-Pr3[0]]
    Pr4f=[Pr4[1],-Pr4[0]]
    #########
    Pt0=[PuntoCfinal[0]+0.5*b*cos(nu-gamma),PuntoCfinal[1]+0.5*b*-sin(nu-gamma)]
    Pt1=[Pt0[0]+(0.5*a)*(-cos(alpha)),Pt0[1]+(0.5*a)*(-sin(alpha))]
    Pt2=[Pt0[0]+(0.5*a+dpbob)*(-cos(alpha)),Pt0[1]+(0.5*a+dpbob)*(-sin(alpha))]
    Pt3=[Pt0[0]+(a+dpbob)*(-cos(alpha)),Pt0[1]+(a+dpbob)*(-sin(alpha))]
    #########################################################################
    ########################################################################
    #########################################################################
    ########################################################################

    d = draw.Drawing(2*R+2*a+2*c+40, 2*R+2*a+2*c+40, origin='center')

    d.append(draw.Circle(0, 0, R+a+c+15,
                fill='white', fill_opacity=0, stroke_width=2, stroke='black'))
    # 
    for i in range(ranuras):
        d.append(draw.ArcLine(0,0,R,angpm+(angxranura*i),(angxranura*i+angxranura-angpm),
                stroke='black', stroke_width=2, fill='white', fill_opacity=0))
    for i in range(ranuras):
        p = draw.Path(stroke_width=2, stroke='black',
                      fill='black', fill_opacity=0)

        p.M(R *cos(angxranura*i*pi/180) - (b/2) * sin(angxranura*i*pi/180),R *sin(angxranura*i*pi/180) + b/2 * cos(angxranura*i*pi/180))
        p.l(c *cos(angxranura*i*pi/180) - 0.5*(anchc-b) * sin(angxranura*i*pi/180),c *sin(angxranura*i*pi/180) + 0.5*(anchc-b) * cos(angxranura*i*pi/180))
        p.l(0 *cos(angxranura*i*pi/180) - (-0.5*(anchc-b)) * sin(angxranura*i*pi/180),0 *sin(angxranura*i*pi/180) + (-0.5*(anchc-b)) * cos(angxranura*i*pi/180))
        p.l(pbc *cos(angxranura*i*pi/180) - 0 * sin(angxranura*i*pi/180),pbc *sin(angxranura*i*pi/180) + 0 * cos(angxranura*i*pi/180))
        p.l(0 *cos(angxranura*i*pi/180) - (-b) * sin(angxranura*i*pi/180),0 *sin(angxranura*i*pi/180) + (-b) * cos(angxranura*i*pi/180))
        p.l(-pbc *cos(angxranura*i*pi/180) - (0) * sin(angxranura*i*pi/180),-pbc *sin(angxranura*i*pi/180) + 0 * cos(angxranura*i*pi/180))
        p.l(0 *cos(angxranura*i*pi/180) - (-0.5*(anchc-b)) * sin(angxranura*i*pi/180),0 *sin(angxranura*i*pi/180) + (-0.5*(anchc-b)) * cos(angxranura*i*pi/180))
        p.l(-c *cos(angxranura*i*pi/180) - 0.5*(anchc-b) * sin(angxranura*i*pi/180),-c *sin(angxranura*i*pi/180) +0.5*(anchc-b) * cos(angxranura*i*pi/180))
        d.append(p)

    p = draw.Path(stroke_width=2, stroke='red',
                      fill='black', fill_opacity=0)
    p.M(PuntoCfinal[0],PuntoCfinal[1])
    p.l(PuntoRfinal[0]-PuntoCfinal[0],PuntoRfinal[1]-PuntoCfinal[1])
    d.append(p)

    d.append(draw.Line(Pa1f[0], Pa1f[1],Pa2f[0],Pa2f[1],fill='red', fill_opacity=0.5, stroke_width=2, stroke='red'))
    d.append(draw.Line(Pa2f[0], Pa2f[1],Pa3f[0],Pa3f[1],fill='red', fill_opacity=0.5, stroke_width=2, stroke='red'))
    d.append(draw.Line(Pa3f[0], Pa3f[1],Pa4f[0],Pa4f[1],fill='red', fill_opacity=0.5, stroke_width=2, stroke='red'))
    d.append(draw.Line(Pa4f[0], Pa4f[1],Pa1f[0],Pa1f[1],fill='red', fill_opacity=0.5, stroke_width=2, stroke='red'))


    d.append(draw.Line(Pr1f[0], Pr1f[1],Pr2f[0],Pr2f[1],fill='red', fill_opacity=0.5, stroke_width=2, stroke='red'))
    d.append(draw.Line(Pr2f[0], Pr2f[1],Pr3f[0],Pr3f[1],fill='red', fill_opacity=0.5, stroke_width=2, stroke='red'))
    d.append(draw.Line(Pr3f[0], Pr3f[1],Pr4f[0],Pr4f[1],fill='red', fill_opacity=0.5, stroke_width=2, stroke='red'))
    d.append(draw.Line(Pr4f[0], Pr4f[1],Pr1f[0],Pr1f[1],fill='red', fill_opacity=0.5, stroke_width=2, stroke='red'))

    d.append(draw.Line(Pt1[0], Pt1[1],Pr3f[0],Pr3f[1],fill='red', fill_opacity=0.5, stroke_width=2, stroke='red'))
    d.append(draw.Line(Pt2[0], Pt2[1],Pa1f[0],Pa1f[1],fill='red', fill_opacity=0.5, stroke_width=2, stroke='red'))
    d.append(draw.Line(Pt3[0], Pt3[1],Pa3f[0],Pa3f[1],fill='red', fill_opacity=0.5, stroke_width=2, stroke='red'))

    n = draw.Path(stroke_width=2, stroke='red',
                      fill='black', fill_opacity=0)
    n.M(PuntoCfinal[0],PuntoCfinal[1])
    n.l(b*cos(nu-gamma),b*-sin(nu-gamma))
    n.l(-(a+dpbob)*cos(alpha),(a+dpbob)*-sin(alpha))
    n.l(b*-cos(nu-gamma),b*sin(nu-gamma))
    n.Z()
    d.append(n)

    d.setPixelScale(2)  # Set number of pixels per geometry unit
    d.setRenderSize(900,900)  # Alternative to setPixelScale
    d.saveSvg('example1.svg')
    #d.savePng('example.png')

    # Display in iPython notebook
    #d.rasterize()  # Display as PNG
    d  # Display as SVG

    #########################################################################
    ########################################################################
    #########################################################################
    ########################################################################

    aa=200
    bb=(200*sin(Phi1))/(sin(Phi2))

    da = draw.Drawing(3*aa,aa, origin='center')

    pa = draw.Path(stroke_width=2, stroke='red',
                      fill='black', fill_opacity=0.5)

    Punto1=[-1.5*aa,0]
    Punto2=[aa*cos(Phi1),aa*sin(Phi1)]
    Punto3=[aa,0]
    Punto4=[bb*cos(Phi2),-bb*sin(Phi2)]
    pa.M(Punto1[0],Punto1[1])
    pa.l(Punto2[0],Punto2[1])
    pa.l(Punto3[0],Punto3[1])
    pa.l(Punto4[0],Punto4[1])
    pa.Z()
    da.append(pa)

    da.setPixelScale(2)  # Set number of pixels per geometry unit
    da.setRenderSize(900,900)  # Alternative to setPixelScale
    da.saveSvg('example2.svg')
    #da.savePng('example2.png')

    # Display in iPython notebook
    #d.rasterize()  # Display as PNG
    #da  # Display as SVG

    #d
    Phi1f.set(Phi1*180/pi)
    Phi2f.set(Phi2*180/pi)
    X1f.set(X1)
    X2f.set(X2)
    Dmaxf.set(distancia)
    
```
